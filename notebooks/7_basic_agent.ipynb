```python
from langchain_core.tools import tool
```

```python
@tool
def add_number(a:int, b:int):
    """
    It will return the sum of two numbers

    Args:
            a : Number one
            b : Number Two
    """
    return a + b


@tool
def multiply_number(a:int, b:int):
    """
    It will return the multiplication of two numbers

    Args:
            a : Number one
            b : Number Two
    """
    return a * b
```

```python
add_number.invoke({"a":12,"b":3})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
agent = create_agent(
    model=llm,
    tools=[add_number, multiply_number],
    system_prompt="you are a math teacher, and always use tool for calculation."
)
```

```python
response = agent.invoke({"messages":[{"role":"user", "content":"what is 2+3"}]})

print(response["messages"][-1].content)
```
